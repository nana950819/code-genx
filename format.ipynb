```python
splitted = s.replace("    ", "\t").split("\n")
splitted = [i for i in splitted if i != ""]
splitted
```

```python
def check_start(element, attributes):
    for att in attributes:
        if element.startswith(att):
            return True
    return False
```

```python
result = []
temp = ""
scope = False
comment_scope = False
for element in splitted:
    if not scope and not comment_scope:
        if check_start(element, ["def", "class"]):
            if temp != "" and not check_start(temp, ["@"]):
                result.append(temp)
                temp = ""
            temp += element
            temp += "\n"
            scope = True
            continue
        elif check_start(element, ["#"]):
            if temp != "":
                result.append(temp)
                temp = ""
            temp += element
            temp += "\n"
            comment_scope = True
            continue
    if scope:
        if check_start(element, ["\t"]):
            temp += element
            temp += "\n"
            continue
        elif check_start(element, ["def", "@", "class"]):
            if check_start(temp, ["@"]):
                temp += element
                temp += "\n"
                continue
            result.append(temp)
            temp = element
            temp += "\n"
            continue
        else:
            scope = False
            result.append(temp)
            temp = element
            temp += "\n"
            continue
    elif comment_scope:
        if check_start(element, ["#", "@", "def", "class"]):
            comment_scope = False
            result.append(temp)
            temp = element
            temp += "\n"
            continue
        else:
            temp += element
            temp += "\n"
            continue
    temp += element
    temp += "\n"

if temp != "":
    result.append(temp)
    temp = ""
```
